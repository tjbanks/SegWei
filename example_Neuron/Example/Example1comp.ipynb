```python
import numpy as np

def activemodel(diam,cm,L,el,gl,gnap,gnas,gnat,gkdr,gca,gh,gkap,gkc,tauca,
                KNap,VhalfNap,KH,VhalfH,KNat,VhalfKdr,KKdr,VhalfNas,KNas,VhalfCa,KCa,gkm,VhalfKm,KKm,VhalfNat,
                tstop,dur,amp,asyne,asyngmax,asynonset,asyntau,
                segkdr,segna3,segnap,seghd,segca,segim,segnas):
    
    ampsave=amp    
    soma.diam = diam
    soma.L = L # um
    soma.cm = cm
    soma.g_pas = gl*1e-6
    soma.e_pas = el
    soma.gbar_kc = gkc*1e-3
    if w_Nas.value:
        soma.gna_nas = gnas*1e-3 
        soma.theta_m_nas = VhalfNas
        soma.sigma_m_nas = KNas
        soma.segnas_nas = segnas
    if w_Nat.value:
        soma.gbar_na3 = gnat*1e-3 
        soma.mhalf_na3 = VhalfNat
        soma.mk_na3 = KNat
        soma.segna3_na3 = segna3
    if w_Nap.value:
        soma.gbar_nap = gnap*1e-6
        soma.k_nap = KNap
        soma.vhalf_nap = VhalfNap
        soma.segnap_nap = segnap
    if w_H.value:
        soma.ghdbar_hd = gh*1e-6
        soma.kl_hd = KH
        soma.vhalfl_hd = VhalfH
        soma.seghd_hd = seghd
    if w_Kdr.value:
        soma.gkdr_kdr = gkdr*1e-3
        soma.theta_hn_kdr = VhalfKdr
        soma.sigma_n_kdr = KKdr
        soma.segkdr_kdr = segkdr
    if w_Cadyn.value:
        soma.gcabar_cadyn = gca*1e-6
        soma.vhalf_cadyn = VhalfCa
        soma.k_cadyn = KCa
        soma.segca_cadyn = segca
    if w_Km.value:
        soma.gbar_im = gkm*1e-4
        soma.vhalf_im = VhalfKm
        soma.k_im = KKm
        soma.segim_im = segim
    if w_Kap.value:
        soma.gkabar_kap = gkap*1e-6
    if w_Capool.value:
        soma.taucas_capool = tauca
    
    if source == 1:
        Source = h.AlphaSynapse(soma(0.5))
        Source.e = asyne
        Source.gmax = asyngmax
        Source.onset = asynonset
        Source.tau = asyntau 
    if source == 0:
        Source = h.IClamp(soma(0.5))
        Source.delay = dur[0]
        Source.dur = dur[1]-dur[0]
        Source.amp = ampsave
    h.tstop = tstop
    h.v_init = el
    
    nc = h.NetCon(soma(0.5)._ref_v,None,sec=soma)
    nc.threshold = 0
    spvec = h.Vector()
    nc.record(spvec)
    passive = []
    passivesaved=[]
    
    h.run()    
      
    print('number of spikes:',len(spvec))

    plt.figure(figsize=(14,10))
    plt.subplot(2,3,5)
    plt.plot(t_vec, v0_vec,'b')
    plt.xlim(0, tstop)
    plt.xlabel('Time (ms)')
    plt.ylabel('Voltage (mV)')
    
    plt.subplot(2,3,3)
    if w_a.value:
        plt.plot(v0_vec,inf_nat,'k.',label = 'minf_Nat')
    if w_b.value:
        plt.plot(v0_vec,inf_nap,'r.',label = 'minf_Nap')
    if w_c.value:
        plt.plot(v0_vec,inf_h,'b.',label = 'hinf_H')
    if w_d.value:
        plt.plot(v0_vec,inf_km,'y.',label = 'ninf_Km')
    if w_e.value:
        plt.plot(v0_vec,inf_kdr,'g.',label = 'ninf_Kdr')
    if w_f.value:
        plt.plot(v0_vec,inf_cadyn,'c.',label = 'uinf_Cadyn')
    if w_g.value:
        plt.plot(v0_vec,inf_nas,'.',label = 'minf_Nas')
    plt.legend()
    plt.xlim(-80,20)
    plt.ylim(0,1)
    plt.xlabel('voltage (mV)')
    plt.ylabel('Inf')
    plt.subplot(2,3,2)
    freq1 = []
    I1 = []
    #if (Source.amp > 0): 
    for Source.amp in np.arange(0,0.8,0.1):
        h.run()
        freq1.append(len(spvec)*1000/Source.dur)
        I1.append(Source.amp)
    plt.xlim(0,1)
    plt.ylim(0,200)
    plt.xlabel('Current Injection (nA)')
    plt.ylabel('Frequency (Hz)')
    for i in range(0, len(frequency)): 
        frequency[i] = float(frequency[i]) 
    for i in range(0, len(amplitude)): 
        amplitude[i] = float(amplitude[i])
    plt.plot(amplitude,frequency,'bo', label = 'Users FIR') 
    plt.plot(I1,freq1,'yo',label = 'FIR')    

    
    plt.legend()
    plt.subplot(2,3,4)   
    if w_a.value:
        plt.plot(v0_vec,tau_nat,'k.',label = 'Nat')
    if w_b.value:
        plt.plot(v0_vec,tau_nap,'r.',label = 'Nap')
    if w_c.value:
        plt.plot(v0_vec,tau_h,'b.',label = 'H')
    if w_d.value:
        plt.plot(v0_vec,tau_km,'y.',label = 'Km')
    if w_e.value:
        plt.plot(v0_vec,tau_kdr,'g.',label = 'Kdr')
    if w_f.value:
        plt.plot(v0_vec,tau_cadyn,'c.',label = 'Ca')
    if w_g.value:
        tau_nas_c=np.zeros(len(v0_vec))
        
        for i in range(len(v0_vec)):
            tau_nas_c[i]=tau_nas
            
        plt.plot(v0_vec,tau_nas_c,'g.',label = 'Nas')
    plt.legend()
    plt.xlim(-80,20)
    plt.ylim(0,2600)
    plt.xlabel('Voltage (mV)')
    plt.ylabel('Tau (mS)')
    plt.subplot(2,3,1)
    
    Source.amp = -0.1
    h.run()
    dt_V = v0_vec[int(dur[0])*40-1000]-v0_vec[int(dur[0])*40+1000]
    R_in = -(dt_V/Source.amp)
    tau = soma.cm*R_in/10
    V_rest = v0_vec[int(tstop)*40-10]
    passive.append(R_in)
    passive.append(tau)
    passive.append(V_rest) 
    print("model's passive property：","Rin:",int(passive[0]),'tau:',int(passive[1]),'Vrest:',int(passive[2]))
    
    print("User's input              ",a.value,b.value,c.value)
    #print(dt_V,v0_vec[int(dur[0])*40-1000],v0_vec[int(dur[0])*40+1000])
    plt.plot(t_vec, v0_vec,'b')
    plt.xlim(0, tstop)
    plt.xlabel('Time (ms)')
    plt.ylabel('Voltage (mV)')
    plt.show()
    
clear_output()
```
